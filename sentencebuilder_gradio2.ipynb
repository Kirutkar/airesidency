```python
import gradio as gr

# Sentence builder logic
def sentence_builder(noun, countries, place, activity_list, time):
    sentence = f"The {noun}s from {' and '.join(countries)} went to the {place} where they {' and '.join(activity_list)} until the {time}."

    grammar_tip = (
        f"This sentence uses a plural subject ('{noun}s') "
        f"and {'multiple actions' if len(activity_list) > 1 else 'a single action'} "
        f"in a descriptive sentence structure."
    )
    return sentence + "\n\nüß† Grammar Tip: " + grammar_tip

# Grammar checker logic
def grammar_check(user_sentence):
    if not user_sentence:
        return "‚ùó Please enter a sentence."
    elif user_sentence[0].islower():
        return "‚ùå Sentence should start with a capital letter."
    elif not user_sentence.endswith("."):
        return "‚ö†Ô∏è Consider ending the sentence with a period (.)"
    else:
        return "‚úÖ Looks grammatically correct!"

# Expanded noun list
nouns = ["üë©‚Äçüè´ Teacher", "üë®‚ÄçüöÄ Astronaut", "üöå Bus", "üöÄ Rocket", "üè† House", "üê∂ Dog", "üê± Cat"]
countries = ["üáÆüá≥ India", "üáØüáµ Japan", "üá´üá∑ France", "üáßüá∑ Brazil", "üá™üá¨ Egypt"]
activities = ["ran", "swam", "ate", "slept", "danced", "painted", "read", "played", "sang"]

with gr.Blocks() as demo:
    gr.Markdown("## ‚ú® Smart Sentence Builder + Grammar Checker\nUse nouns, actions, and grammar tips to build creative English sentences!")

    with gr.Row():
        with gr.Column():
            noun = gr.Dropdown(nouns, label="Noun (Person/Thing)")
            countries_input = gr.CheckboxGroup(countries, label="Countries")
            place = gr.Radio(["park", "zoo", "road", "garden", "beach"], label="Location")
            activities_input = gr.Dropdown(activities, multiselect=True, label="Activities")
            time = gr.Radio(["morning", "afternoon", "evening", "night"], label="Time of Day")
            submit_btn = gr.Button("Build Sentence")

        with gr.Column():
            output = gr.Textbox(label="Generated Sentence with Grammar Tip", lines=5)

    submit_btn.click(
        fn=sentence_builder,
        inputs=[noun, countries_input, place, activities_input, time],
        outputs=output
    )

    gr.Markdown("---\n### ‚úçÔ∏è Check Your Own Sentence Below")

    user_input = gr.Textbox(label="Write Your Own Sentence")
    grammar_btn = gr.Button("Check Grammar")
    grammar_output = gr.Textbox(label="Grammar Feedback")

    grammar_btn.click(grammar_check, user_input, grammar_output)

if __name__ == "__main__":
    demo.launch()
```
